### Checking the missing values

```python
tour_data.isna().sum().sort_values(ascending = False)
```

### Handling the missing values

```python
tour_data['Actual Price'] = tour_data['Actual Price'].fillna(tour_data['Actual Price'].mean())
tour_data['Hotel Rating'] = tour_data['Hotel Rating'].fillna(tour_data['Hotel Rating'].median())
tour_data['Discount (%)'] = tour_data.apply(
    lambda row: ((row['Actual Price'] - row['Offer Price']) / row['Actual Price']) * 100
    if pd.isnull(row['Discount (%)']) and pd.notnull(row['Actual Price']) and pd.notnull(row['Offer Price'])
    else row['Discount (%)'],
    axis=1
)
```

```python
tour_data.isna().sum().sort_values(ascending = False)
```

```python
tour_data = tour_data.dropna(subset=['Cities'])
```

```python
tour_data.isna().sum().sort_values(ascending = False)
```

### Handling duplicates

```python
duplicate_rows = tour_data[tour_data.duplicated(keep=False)]
duplicate_rows
```

```python
tour_data = tour_data.drop_duplicates()
```

```python
tour_data.shape
```

### Data Type Conversion

```python
tour_data["Actual Price"] = tour_data["Actual Price"].astype('int64')
tour_data["Discount (%)"] = tour_data["Discount (%)"].astype('int64')
tour_data["Hotel Rating"] = tour_data["Hotel Rating"].astype('int64')
tour_data["Offer Price"] = tour_data["Offer Price"].astype('int64')
tour_data.info()
```

```python
#Which cities are most frequently covered in the packages?

import re

# Extract all city names using regex before the duration (e.g., 'Goa' from 'Goa (3D)')
def extract_cities(text):
    return re.findall(r'([A-Za-z\s]+)\s*\(\d+D\)', text)

# Apply the extraction and explode the list into rows
cities_series = tour_data["Cities"].apply(extract_cities).explode()

# Get the top 10 most common cities
most_common_cities = cities_series.value_counts().head(10)

print(most_common_cities)
```

```python
# Find outliers in the old price and new price using the IQR method.
def find_outliers(column):
    Q1 = tour_data[column].quantile(0.25)
    Q3 = tour_data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = tour_data[(tour_data[column] < lower_bound) | (tour_data[column] > upper_bound)]
    return outliers

old_price_outliers = find_outliers('Actual Price')
new_price_outliers = find_outliers('Offer Price')
{"Actual Price Outliers": len(old_price_outliers), "Offer Price Outliers": len(new_price_outliers)}
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import color_palette
import re

# Step 1: Extract cities from 'Cities' column
def extract_cities(text):
    return re.findall(r'([A-Za-z\s]+)\s*\(\d+D\)', str(text))

# Step 2: Create exploded DataFrame
tour_data["Extracted Cities"] = tour_data["Cities"].apply(extract_cities)
exploded_df = tour_data.explode("Extracted Cities").copy()
exploded_df["Extracted Cities"] = exploded_df["Extracted Cities"].str.strip()

# Step 3: Compute average Price Per Day for each city
city_price_stats = (
    exploded_df.groupby("Extracted Cities")["Price Per Day"]
    .mean()
    .sort_values(ascending=False)
)

# Step 4: Plotting
# Create reversed color palettes
red_colors = color_palette("Reds", 5)[::-1]     # Darker red for more expensive
green_colors = color_palette("Greens", 5)[::-1] # Lighter green for cheaper cities

plt.figure(figsize=(12, 6))

# Top 5 most expensive cities
plt.subplot(1, 2, 1)
sns.barplot(x=city_price_stats.head(5).index,
            y=city_price_stats.head(5).values,
            palette=red_colors)
plt.title("Top 5 Most Expensive Cities (Per Day)")
plt.xlabel("City")
plt.ylabel("Avg Price Per Day")
plt.xticks(rotation=45)

# Bottom 5 cheapest cities
plt.subplot(1, 2, 2)
sns.barplot(x=city_price_stats.tail(5).index,
            y=city_price_stats.tail(5).values,
            palette=green_colors)
plt.title("Top 5 Cheapest Cities (Per Day)")
plt.xlabel("City")
plt.ylabel("Avg Price Per Day")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()


```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Avoid division by zero
tour_data = tour_data[tour_data['Duration (Days)'] > 0].copy()

# Calculate price per day for each row
tour_data['Price Per Day'] = tour_data['Offer Price'] / tour_data['Duration (Days)']

# Group by duration and calculate average price per day
avg_price_per_day = tour_data.groupby('Duration (Days)')['Price Per Day'].mean().reset_index()

# Plot
plt.figure(figsize=(10, 6))
plt.plot(
    avg_price_per_day['Duration (Days)'],
    avg_price_per_day['Price Per Day'],
    color='teal',
    marker='o',
    linewidth=2
)

# Labels and styling
plt.title('Average Offer Price per Day vs Duration', fontsize=14)
plt.xlabel('Duration (Days)', fontsize=12)
plt.ylabel('Avg Price per Day (₹)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

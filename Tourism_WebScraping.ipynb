```python
import requests
from bs4 import BeautifulSoup
import numpy as np

package_name = []
durations = []
cities = []
old_prices = []
new_prices = []
discounts = []
ratings = []

for j in range(1, 155):  # Loop through pages 1 to 155
    mu_url = f"https://traveltriangle.com/tour-packages/india/page/{j}"
    page = requests.get(mu_url)
    print(f"Fetching: {mu_url} -> Status: {page.status_code}")
    
    soup = BeautifulSoup(page.text, "html.parser")
    packages = soup.find_all("div", class_="col-12 p8 pt0")  # each package container

    for package in packages:
        # Package name
        name = package.find("span", class_="fw9 m0 f16 pfc3")
        package_name.append(name.text.strip() if name else np.nan)

        # Duration
        dur = package.find("span", class_="iblock sfc6")
        durations.append(dur.text.strip() if dur else np.nan)

        # Cities covered
        city_list = package.find("ul", class_="clearfix package-cities-list at_packagecity_list")
        cities.append(city_list.text.strip() if city_list else np.nan)

        # Old price
        old = package.find("span", class_="f12 pfc3 tdl ml8 at_oldprice fw4")
        old_prices.append(old.text.strip() if old else np.nan)

        # New price (avoid pulling old price from nested span)
        p_tag = package.find("p", class_="priceVal")
        if p_tag:
            price_text = p_tag.find(text=True, recursive=False)
            new_prices.append(price_text.strip() if price_text else np.nan)
        else:
            new_prices.append(np.nan)

        # Discount
        disc = package.find("span", class_="f12 fw4 pt2 pb2 pl8 pr8 radius20 pbc1 sfcw at_discount_label")
        discounts.append(disc.text.strip() if disc else np.nan)

        # Hotel star rating
        rating_input = package.find("input", {"class": "radio-common-circle at_packagehotel_rating", "checked": True})
        label = rating_input.find_next("label") if rating_input else None
        ratings.append(label.text.strip().split(" ")[0] if label else np.nan)

# Optionally convert some lists to NumPy arrays
old_prices = np.array(old_prices, dtype=object)
discounts = np.array(discounts, dtype=object)

# Print summary lengths to verify all aligned
print("Packages:", len(package_name))
print("Durations:", len(durations))
print("Cities:", len(cities))
print("Old Prices:", len(old_prices))
print("New Prices:", len(new_prices))
print("Discounts:", len(discounts))
print("Ratings:", len(ratings))
```

```python
df = pd.DataFrame({"Package Name":package_name,"Durations":durations,"Cities Covered":cities,"Actual Price":old_prices,"Offer Price":new_prices,"Discount(%)":discounts,"Hotel_rating":ratings})
df
```

```python
def extract_duration(durations):
    days = []
    nights = []
    for duration in durations:
        day_match = re.search(r"(\d+)\s*Days?", duration)
        night_match = re.search(r"(\d+)\s*Nights?", duration)
        days.append(int(day_match.group(1)) if day_match else None)
        nights.append(int(night_match.group(1)) if night_match else None)
    return days, nights


days, nights = extract_duration(durations)
print("Days:", days)
print("Nights:", nights)
print("Length of Days:", len(days))
print("Length of Nights:", len(nights))
```

```python
def clean_price(prices):
    clean_prices = []
    for price in prices:
        if isinstance(price, (int, float)):
            price = str(price)
        elif not isinstance(price, str):
            clean_prices.append(None)
            continue
        
        match = re.search(r"(\d{1,3}(,\d{3})*|\d+)", price)
        if match:
            clean_prices.append(int(match.group(1).replace(",", "")))
        else:
            clean_prices.append(None)
    
    return clean_prices


old_prices_cleaned = clean_price(old_prices)
new_prices_cleaned = clean_price(new_prices)
print("Old Prices Cleaned:", old_prices_cleaned)
print("New Prices Cleaned:", new_prices_cleaned)
print(f"Old_prices :{len(old_prices_cleaned)}")
print(f"New_prices :{len(new_prices_cleaned)}")
```

```python
discounts_cleaned = [
    int(re.sub(r"[^\d]", "", str(discount))) if isinstance(discount, (str, int, float)) and re.search(r"\d", str(discount)) 
    else None 
    for discount in discounts
]
```

```python
data = pd.DataFrame({"Package Name": package_name,"Duration (Days)": days,"Duration (Nights)": nights,"Cities":cities,"Actual Price": old_prices_cleaned,"Offer Price": new_prices_cleaned,"Discount (%)": discounts_cleaned,"Hotel Rating":ratings})
data
```

```python
data.to_csv("India_Tours_Packages.csv")
```
